```python
# Import Numpy for calculations and matplotlib for charting

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


%matplotlib inline
```

---

#### Iris Data

```python
# Load a dataset

iris = sns.load_dataset('iris')
```

```python
# Print a preview sample data

iris[:5].T
```

```python
iris.species.unique()
```

```python
# Almost the same as the Pairgrid, but you have more customization control on the pairgrid than pairplot.

sns.pairplot(iris, palette='coolwarm', hue='species', kind='scatter')
```

#### PairGrid

```python
# Plot the Data using Pairgrid
# Using Pairgrid, you can customize the individual elements
# i.map_diag, i.map_upper, i.map_lower, i.map_offdiag

i = sns.PairGrid(iris)
i.map_diag(plt.hist)
i.map_lower(sns.kdeplot)
i.map_upper(sns.scatterplot)


# sns.PairGrid(iris).map_diag(plt.hist).map_lower(sns.kdeplot).map_upper(sns.scatterplot)
```

```python
# Load a dataset

tips = sns.load_dataset('tips')
```

```python
tips[:5].T
```

#### FaceGrid

```python
sns.FacetGrid(tips, row='time', col='smoker', hue='sex').map(plt.hist, 'total_bill')
```

```python
sns.FacetGrid(tips, row='time', col='smoker', hue='sex').map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
sns.FacetGrid(tips, row='sex', col='smoker', hue='time').map(plt.scatter, 'total_bill', 'tip').add_legend()
```

---

#### Joint Grid

```python
sns.JointGrid(x='total_bill', y='tip', data=tips).plot(sns.regplot, sns.distplot)
```

#### CountPlot

```python
# Apply Style

#sns.set_style('whitegrid')

sns.set_style('darkgrid')
sns.countplot(x='time', data=tips)
sns.despine(left=True)              
```

```python
# Apply Style

#sns.set_style('whitegrid')

plt.figure(figsize=(12,6), dpi=300)
sns.countplot(x='time', data=tips)
    
```
