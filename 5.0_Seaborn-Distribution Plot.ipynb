```python
# Import Numpy for calculations and matplotlib for charting

import numpy as np
import seaborn as sns


%matplotlib inline
```

---

#### Distribution Plot

```python
# Load a dataset

tips = sns.load_dataset('tips')
```

```python
# Print a preview sample data

tips[:5].T
```

```python
# Plot distribution for tip

sns.distplot(tips['tip'])
```

```python
# Plot distribution for ``total_bill``

sns.distplot(tips['total_bill'],kde=False)
```

```python
# Plot kde distribution for ``total_bill`` 

sns.distplot(tips['total_bill'],hist=False)
```

```python
# Plot distribution for ``total_bill`` 

sns.distplot(tips['total_bill'],kde=False, bins=40)
```

```python
# Plot kde distribution for ``tip`` 

sns.distplot(tips['tip'],kde=True, hist=False, rug=True, bins=40)
```

---

#### Joint Plot

```python
# Scatter Plot

sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips)
```

```python
# Kde Plot

sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind='kde')
```

```python
# Hex Plot

sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind='hex')
```

```python
# Regression Model Plot

sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind='reg')
```

---

#### PairPlot

> This is very useful in understanding your data.

> This will plot all the numerical columns of your dataset and plot against each individual variable

> It will show correlation between the variables of your data

```python
sns.pairplot(tips)
```

```python
# comparison between the customer gender


sns.pairplot(tips, hue='sex',kind='reg')
```

```python
# comparison between the smoker customer or non-smoker


sns.pairplot(tips, hue='smoker',kind='reg')
```

```python
# comparison between the dining time (launch or dineer)


sns.pairplot(tips, hue='time',kind='scatter', palette='coolwarm')
```

```python
# comparison between the dining time (launch or dineer)
sns.kdeplot(tips['tip'])
sns.kdeplot(tips['total_bill'])
```
