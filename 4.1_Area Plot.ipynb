```python
# Import Numpy for calculations and matplotlib for charting

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#prettify the plot
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
# Create Synthetic Data

df = pd.DataFrame(np.random.randint(100,size=(10,4)), columns=['a','b','c','d'])
df.head()
```

```python
# Area Plot - Multi-Column
plt.figure()
df.plot.area()
plt.savefig('Resources/pltsave/areaplot.png', bbox_inches='tight')
```
