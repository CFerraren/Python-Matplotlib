```python
# Import Numpy for calculations and matplotlib for charting

import numpy as np
import matplotlib.pyplot as plt

#prettify the plot
import matplotlib
matplotlib.style.use('ggplot')
```

```python
# Create our x_axis list

x_axis = np.arange(0, 6, 0.1)
```

```python
# Print the Synthetic generated number with a range from 0 < 6, step_size = 0.1

x_axis
```

```python
# Creates a list based on the sin of our x_axis values

sin = np.sin(x_axis)
```

```python
# Creates a list based on the cos of our x_axis values

cos = np.cos(x_axis)
```

#### Plot

```python
# Plot both of these lines so that they will appear on our final chart


plt.plot(x_axis, sin)
plt.plot(x_axis, cos)

plt.show()
```
