```python
# Import Numpy for calculations and matplotlib for charting

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Creat an array of 10 equally distant numbers

x= np.linspace(0,9,10)
x
```

```python
y = x ** 3
y
```

```python
# plot the variable x and y

plt.plot(x,y, color='green', marker='o', linestyle='dashed',
        linewidth=0.5, markersize=5)

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title of Plot')
```

#### Multiple Plot Subjections

```python
# Multiple Subplots
#Create 2x2 subplots

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x, '--r')
plt.subplot(2,2,2)
plt.subplot(2,2,4)
```

---

#### Api Functionality

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, 'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')

#-----

axes1 = fig.add_axes([0.1,0.4,0.5,0.5])
axes1.plot(y,x, 'b--')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Plot Title')
```

---

```python
fig, axes = plt.subplots()
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')
axes[0].set_title('1st Plot Title')


axes[1].plot(y,x, '--r')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
axes[1].set_title('2nd Plot Title')


fig.tight_layout
```

---

```python
fig = plt.figure(figsize=(10,4), dpi=100)

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')
```

```python
fig, axes = plt.subplots(figsize=(6,6))
axes.plot(x,y, 'r--')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plot Title')

#print the figure
fig
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, 'r--', label='x**2')
axes.plot(x, x**3, 'b--', label='x**3')
axes.legend(loc=10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, 'rv', label='x square')
axes.plot(x, x**3, 'b^', label='x cube')
axes.legend(loc=5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, 'r-', linewidth=0.5, label='x square',marker='+')
axes.plot(x, x**3, color='#000000', lw=1, label='x cube', marker='o')
axes.legend(loc=5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**3, x, x**2)
axes.set_xlim(2,8)
axes.set_ylim(10,300)
```
