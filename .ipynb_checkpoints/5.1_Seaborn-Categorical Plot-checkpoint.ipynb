```python
# Import Numpy for calculations and matplotlib for charting

import numpy as np
import seaborn as sns


%matplotlib inline
```

---

#### Categorical Plot

```python
# Load a dataset

tips = sns.load_dataset('tips')
```

```python
# Print a preview sample data

tips[:5].T
```

```python
# Tipping habits of sexes between different dining times 

sns.barplot(x=tips.time, y=tips.tip, hue='sex', data=tips)
```

```python
# Plot to find the big spenders between sexes

sns.barplot(x=tips.time, y=tips.total_bill, hue='sex', data=tips)
```

```python
# Plot smoker customers spending habits compared to non-smoker

sns.barplot(x=tips.sex, y=tips.total_bill, hue='smoker', data=tips)
```

```python
# Plot of tipping habits between smoker and non-smoker

sns.barplot(x=tips.sex, y=tips.tip, hue='smoker', data=tips)
```

```python
# Plot tipping between different days of the week

sns.barplot(x=tips.sex, y=tips.tip, hue='day', data=tips)
```

```python
# Plot of spending between different days of the week

sns.barplot(x=tips.sex, y=tips.total_bill, hue='day', data=tips)
```

---

#### CountPlot

```python
# Plot of tipping between sexes

sns.countplot(x=tips.sex, data=tips)
```

```python
# Plot of between sexes on dining time

sns.countplot(x=tips.sex, hue=tips.time, data=tips, palette='coolwarm')
```

---

#### Boxplot

```python
# Plot of between sexes on spending for diner

sns.boxplot(x='sex', y='total_bill', data=tips, palette='coolwarm')
```

```python
# Plot of between dining time

sns.boxplot(x='time', y='total_bill', data=tips, palette='coolwarm')
```

```python
ax = sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
# Plot of between spending on dining days

sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm')
```

```python
# Plot of between spending habits of smoking and non-smoking customers

sns.boxplot(x='sex', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
```

```python
# Plot all the numerical data

sns.boxplot(data=tips)
```

```python
# Plot all the numerical data

sns.boxplot(data=tips, orient='h')
```

```python
# Plot of between spending habits of smoking and non-smoking customers

sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
```

---

#### Violin Plot

```python
# Spending between days across smoker and non-smoker

sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
```

```python
# Tipping between days across smoker and non-smoker

sns.violinplot(x="day", y="tip", hue="smoker", data=tips, palette="muted")
```

```python
# Tipping between days across sexes

sns.violinplot(x="day", y="tip", hue="sex", data=tips, palette="muted")
```

```python
# Tipping between days across sexes splitted

sns.violinplot(x="day", y="tip", hue="sex", data=tips, palette="muted", split=True)
```

---

#### Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True, palette='deep')
```

#### Swarmplot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.swarmplot(x='total_bill', y='day', data=tips, hue='sex', dodge=True)
sns.violinplot(x='total_bill', y='day', data=tips, palette='rainbow')
```

---

#### Factor Plot -> Catplot

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='point')
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='box')
```
