```python
# Import Libraries

import os
import pandas as pd
import numpy as np


%matplotlib inline
```

```python
# Location

file = os.path.join('Resources', 'AAPL.csv')
```

```python
# Read csv
# Parse date to datetime format

aapl = pd.read_csv(file, parse_dates=['Date'])
aapl.head()
```

#### Check Data Types

```python
aapl.dtypes
```

```python
# Reverse the dataframe to show latest date

aapl=aapl[::-1]
aapl.head()
```

#### Record per Day

```python
#Set the index to Date

aapl.set_index('Date', inplace=True)
aapl.head()
```

#### Summarized per Month

```python
# Summarizing Stocks by Month

df = aapl.resample('M').mean()
df
```

#### Summarized per Quarter

```python
# Summarizing Stocks by Quarter

dfq = aapl.resample('Q').mean()
dfq
```

---

#### Bar Plot

```python
# per Day

aapl['Adj Close'].plot(kind='bar')
```

> Suitalbe for low data granularity. Here, 'Date' is the x-Axis.

```python
# per Month

df['Adj Close'].plot(kind='bar')
```

```python
# per Quarter

dfq['Adj Close'].plot(kind='bar')
```
